#Descrição:

Você foi contratado para desenvolver um sistema básico de gerenciamento de funcionários, através de um banco de dados temporário, para uma pequena empresa. O sistema deve ser capaz de receber e armazenar informações sobre os funcionários e realizar operações básicas de manipulação desses dados.
Requisitos:
Utilize a função input() para realizar um cadastro coletando dados do usuário, como nome, cidade e estado em que reside, idade, escolaridade, cargo e salário.

#Funcionalidades do Sistema:

O sistema deve oferecer as seguintes funcionalidades:
Listar todos os funcionários cadastrados.
Exibir informações de um funcionário específico.
Adicionar um novo funcionário ao sistema, solicitando informações como nome, cargo, salário, e-mail, etc.
Atualizar as informações de um funcionário existente com base em um identificador único (como um número de identificação, cpf, index ou nome).
Permitir a remoção de um funcionário do sistema.

#IMPORTANTE!!!

1. Antes de rodar o código GERENCIADOR BANCO DE DADOS, rodar o código CARGA INICIAL.
2. Para sair do programa, digitar a operação 6, de acordo com o menu principal.
3. Nas perguntas de resposta Sim=1 e Não=2, qualquer resposta diferente de 1 vai retornar o menu principal.

```python
#CARGA INICIAL

cpfs=[111111, 222222, 333333,444444,555555]
nomes=['jose','joao','jose','pedro','joana']
cidades=['sjc','belo horizonte','curitiba','manaus','recife']
estados=['SP','MG','PR','am','PE']
idades=[55,32,99,45,22]
escolaridades=['superior','medio','fundamental','maternal','jd de infancia']
cargos=['gerente','supervisor','analista','porteiro','copeiro']
salarios=[15000,10000,50000,3000,1500]
```

```python
#GERENCIADOR BANCO DE DADOS

while True:
    print('\n')
    print('OPERAÇÕES:')
    print('1 - Listar funcionarios')
    print('2 - Buscar dados de funcionario')
    print('3 - Adicionar novo funcionario')
    print('4 - Atualizar dados de funcionario')
    print('5 - Remover funcionario')
    print('6 - SAIR')
    print('\n')
    operacao=int(input('Digite o numero da operação desejada: '))
    print('\n')

    #LISTAR FUNCIONARIOS
    if operacao==1:
      for i in range(len(nomes)):
        print('CPF:',cpfs[i],' / ','Nome:',nomes[i].strip().title(),' / ','Cidade:',cidades[i].strip().title(),' / ','Estado:',estados[i].strip().upper(),' / ','Idade:',idades[i],' / ','Escolaridade:',escolaridades[i].strip().title(),' / ','Cargo:',cargos[i].strip().title(),' / ','Salario:',salarios[i])

    #BUSCAR DADOS DE FUNCIONARIOS
    elif operacao==2:
      aviso=1
      while aviso==1:
        x=int(input('Insira CPF: '))
        if x in cpfs:
          ind=cpfs.index(x)
          print('CPF:',cpfs[ind],' / ','Nome:',nomes[ind].strip().title(),' / ','Cidade:',cidades[ind].strip().title(),' / ','Estado:',estados[ind].strip().upper(),' / ','Idade:',idades[ind],' / ','Escolaridade:',escolaridades[ind].strip().title(),' / ','Cargo:',cargos[ind].strip().title(),' / ','Salario:',salarios[ind])
        else:
          print('Funcionário não encontrado.')
        aviso=int(input('Deseja fazer nova consulta? (Sim=1;Não=2)'))
        if aviso!=1:
          break

    #ADICIONAR NOVO FUNCIONARIO
    elif operacao==3:
      aviso=1
      while aviso==1:
        cpf=int(input('Inserir CPF do novo funcionario (XXXXXX): '))
        if len(str(cpf))!=6:
          print('Formato invalido')
        else:
          if cpf in cpfs:
            print('CPF já existente.')
          else:
            cpfs.append(cpf)
            nome=input('Inserir nome: ')
            nomes.append(nome)
            cidade=input('Inserir cidade: ')
            cidades.append(cidade)
            estado=input('Inserir estado: ')
            estados.append(estado[:2])
            idade=int(input('Inserir idade: '))
            idades.append(idade)
            escolaridade=input('inserir escolaridade: ')
            escolaridades.append(escolaridade)
            cargo=input('Inserir cargo: ')
            cargos.append(cargo)
            salario=float(input('Inserir salário: '))
            salarios.append(salario)
          aviso=int(input('Deseja inserir outro funcionario? (Sim=1;Não=2)'))
        if aviso!=1:
            #print('----------------------------\n')
            for i in range(len(nomes)):
              print('CPF:',cpfs[i],' / ','Nome:',nomes[i].strip().title(),' / ','Cidade:',cidades[i].strip().title(),' / ','Estado:',estados[i].strip().upper(),' / ','Idade:',idades[i],' / ','Escolaridade:',escolaridades[i].strip().title(),' / ','Cargo:',cargos[i].strip().title(),' / ','Salario:',salarios[i])
            #print('----------------------------\n')
            break

    #ATUALIZAR DADOS DE FUNCIONARIO
    elif operacao==4:
      aviso1=1
      while aviso1==1:
        x=int(input('entre com o cpf que quer atualizar: '))
        aviso=1
        while aviso==1:
          if x in cpfs:
            ind=cpfs.index(x)
            print('10 - cidade/estado')
            print('30 - escolaridade')
            print('40 - cargo')
            print('50 - salario\n')
            item=int(input("Qual informação deseja atualizar?\n"))
            if item==10:
              city=input('entre com a nova cidade: ')
              state=input('entre com o novo estado: ')
              cidades[ind]=city
              estados[ind]=state
            elif item==30:
              scholar=(input('entre com a nova escolaridade: '))
              escolaridades[ind]=scholar
            elif item==40:
              pos=input('entre com o novo cargo: ')
              cargos[ind]=pos
            elif item==50:
              pay=int(input('entre com o novo salario: '))
              salarios[ind]=pay
            else:
              print('Item desconhecido')
            for i in range(len(nomes)):
              print('CPF:',cpfs[i],' / ','Nome:',nomes[i].strip().title(),' / ','Cidade:',cidades[i].strip().title(),' / ','Estado:',estados[i].strip().upper(),' / ','Idade:',idades[i],' / ','Escolaridade:',escolaridades[i].strip().title(),' / ','Cargo:',cargos[i].strip().title(),' / ','Salario:',salarios[i])
          else:
            print('CPF desconhecido')
          aviso=int(input(f'Deseja atualizar outra informação no CPF {x} (Sim=1;Não=2)'))
          if aviso!=1:
            break
        aviso1=int(input('Deseja atualizar outro CPF? (Sim=1;Não=2)'))
        if aviso1!=1:
          break

    #REMOVER FUNCIONARIO
    elif operacao==5:
      aviso=1
      while aviso==1:
        x=int(input('entre com o cpf: '))
        if x in cpfs:
          ind=cpfs.index(x)
          cpfs.pop(ind)
          nomes.pop(ind)
          cidades.pop(ind)
          estados.pop(ind)
          idades.pop(ind)
          escolaridades.pop(ind)
          cargos.pop(ind)
          salarios.pop(ind)
        else:
          print('CPF desconhecido')
        aviso=int(input('Deseja remover outro CPF? (Sim=1;Não=2)'))
        if aviso!=1:
          for i in range(len(nomes)):
            print('CPF:',cpfs[i],' / ','Nome:',nomes[i].strip().title(),' / ','Cidade:',cidades[i].strip().title(),' / ','Estado:',estados[i].strip().upper(),' / ','Idade:',idades[i],' / ','Escolaridade:',escolaridades[i].strip().title(),' / ','Cargo:',cargos[i].strip().title(),' / ','Salario:',salarios[i])
          break

    #SAIR
    elif operacao==6:
      break

    #OPERAÇÃO NÃO RECONHECIDA
    else:
      print('Operação não reconhecida')

```
