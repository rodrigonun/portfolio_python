#1. Importar a base de dados

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

tabela=pd.read_csv(r'G:\Meu Drive\2_REPOSITORIOS\PYTHON\PTH_009_IA_NOTA_CREDITO\clientes.csv')
```

#2. Preparação da base de dados para a IA

###2.1. Verificando os tipos das colunas

```python
tabela.info()
tabela.head()
```

```python
profissao=tabela['profissao'].unique()
for i in profissao:
    print(i)
```

```python
comportamento_pagamento=tabela['comportamento_pagamento'].unique()
for i in comportamento_pagamento:
    print(i)
```

```python
mix_credito=tabela['mix_credito'].unique()
for i in mix_credito:
    print(i)
```

```python
score_credito=tabela['score_credito'].unique()
for i in score_credito:
    print(i)
```

###2.2. Transformação de coluna texto (formato 'object') em numeros inteiros

```python
#Transformar colunas de texto em numero inteiro
codificador_prof=LabelEncoder()
tabela['profissao']=codificador_prof.fit_transform(tabela['profissao'])

codificador_mc=LabelEncoder()
tabela['mix_credito']=codificador_mc.fit_transform(tabela['mix_credito'])

codificador_cp=LabelEncoder()
tabela['comportamento_pagamento']=codificador_cp.fit_transform(tabela['comportamento_pagamento'])

tabela.info()
tabela.head(10)
```

###2.3. Separação das informações da base de dados para a IA

```python
#Separar as informções em X e Y
# X são as outras colunas que eu usarei para fazer a previsão, os valores dessas colunas devem ser tipo int or float (NUMÉRICOS)
# Y é a coluna que eu quero prever (coluna SCORE_CREDITO), os valores dessa coluna pode ser de qualquer tipo

x=tabela.drop(columns=['score_credito','id_cliente'])
y=tabela['score_credito']
```

```python
#Separar em treino e teste
from sklearn.model_selection import train_test_split

#essa ordem deve sempre ser obedecida
x_treino,x_teste,y_treino,y_teste = train_test_split(x,y,test_size=0.3) # padrão é teste=30% e treino=70%
```

#3. Criar o modelo de IA => prever nota de crédito

Passos para criar um modelo de IA
1. Importar o modelo
2. Criar o modelo
3. Treinar o modelo

Modelos escolhidos para esse projeto:
- Modelo1: Arvore de decisão (RamdonForest)
- Modelo2: Vizinhos próximos (KNN)

```python
#1. Importando os modelos
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
#2. Criando os modelos
modelo_arvoredecisao=RandomForestClassifier()
modelo_vizinhosprox=KNeighborsClassifier()
```

```python
#3. Treinar os modelos
modelo_arvoredecisao.fit(x_treino,y_treino)
modelo_vizinhosprox.fit(x_treino,y_treino)
```

#4. Escolher o melhor modelo

```python
#Previsões
previsao_arvoredecisao=modelo_arvoredecisao.predict(x_teste)
previsao_vizinhosprox=modelo_vizinhosprox.predict(x_teste)
```

```python
#Acuracia (precisão de cada modelo)
from sklearn.metrics import accuracy_score

display(accuracy_score(y_teste,previsao_arvoredecisao))
display(accuracy_score(y_teste,previsao_vizinhosprox))
```

###Conclusão
De acordo com os resultados, o modelo que melhor prevê os parâmetro desejados é o de árvore de decisão (82%)

###Como melhorar a precisão do modelo?
1. Testar outros modelos => pode ser que outros modelos possam se ajustar melhor a base de dados;
2. Treinar a IA com uma base de dados maior => ou conseguir mais dados ou aumentar a porcentagem de treino (nunca maior q 90%)
3. Alterar os parâmetros da IA
